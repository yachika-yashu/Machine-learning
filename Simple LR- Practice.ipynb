```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df=pd.read_csv('insurance.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X=df.iloc[:,0:1]
y=df.iloc[:,-1]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
plt.scatter(df['age'],df['charges'])
plt.plot(X_train,lr.predict(X_train), color='red')
plt.title('age vs charges')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
y_pred=lr.predict(X_test)
```

```python
MAE=mean_absolute_error(y_test,y_pred)
print(MAE)
MSE=mean_squared_error(y_test,y_pred)
print(MSE)
R2_Score=r2_score(y_test,y_pred)
print(R2_Score)
#Adjusted R2
ar2=1-(1-R2_Score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print('Adjusted R2 Score:',ar2)
```

Model Evaluation Summary
🔹 MAE = ₹9,147.18
→ On average, predictions are off by ₹9K+ — quite high. Poor predictive accuracy.

🔹 MSE = 139,176,465.01
→ Very high! Indicates some predictions are way off. MSE exaggerates big errors due to squaring — possibly outliers.

🔹 R² = 0.1254
→ Model explains only 12.5% of the variance. Not good — ~87.5% remains unexplained. because there are other parameters as well in the data.

🔹 Adjusted R² = 0.1221
→ Slightly lower than R², meaning model isn’t overfitting but still performs poorly.

📉 Conclusion: Model needs improvement — likely underfitting or missing strong predictors.
