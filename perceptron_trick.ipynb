Uding make_classification, generated  a dataset with 100 samples, 2 columns and 2 classes of classification and then plot that dataser

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

made a new function perceptron to which if you give dataset it will return intercepts and coefficients of the line

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1) # now x contains a column of 1's that is X0, x1 and x2
    weights = np.ones(X.shape[1]) # here w contains all 1 in shape of X ie w0,w1,w2 here all 1
    lr = 0.1 # learning rate

    for i in range(1000): # epoch=100
        j = np.random.randint(0,100) #select a random data point if j comes out 41 we will pick 41st student
        y_hat = step(np.dot(X[j],weights)) # dot product of weights(w0,w1,w2) and x[41] which is (x0,x1,x2) gives prediction and then applying step function will give classification
        weights = weights + lr*(y[j]-y_hat)*X[j] # equation from algorithm to find w0,w1,w2 and so on based on the point you choose, line equation move wrt point chosen

    return weights[0],weights[1:]

def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
print(coef_)
print(intercept_)
```

```python

m = -(coef_[0]/coef_[1]) #slope
b = -(intercept_/coef_[1]) #
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b #line equation
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
